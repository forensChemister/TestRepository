```python
class Patient:
    
    temp_list = []
    temp = 36.6

    def __init__(self, n, ds,  bd) -> None:
        self.__name = n
        self.__diagnosis = ds
        self.__birth_date = bd
        
    
    def set_name(self):
        self.__name = input()
     
    def get_name(self):
        print(self.__name)
    

    def set_diagnosis(self):
        self.__diagnosis = input()
    

    def get_diagnosis(self):
        print(self.__diagnosis)

    def set_birth_date(self):
        self.__birth_date = input()
    

    def get_birth_date(self):
        print(self.__birth_date)
    

    def set_temp(self):
       Patient.temp = float(input())
       while Patient.temp > 0:
            Patient.temp_list.append(Patient.temp)
            Patient.temp = float(input())
   

    def get_temp(self):
        print(Patient.temp_list[-1])
    

    def avg_temp (self):
        return sum(Patient.temp_list) / len(Patient.temp_list)
    
    def calc_age(self):
        today = DT.date.today()
        birthsday_date = DT.datetime.strptime(self.__birth_date, '%Y%m%d').date()
        age = (today - birthsday_date) // DT.timedelta(days=365)
        print(age)
    

    def draw_temp_curve(self):
        y = Patient.temp_list
        x = [i+1 for i in range(len(y))]
        plt.figure(figsize=(9, 5))
        plt.plot(x, y, marker="o", c="r")
        plt.xlabel('Дни') #Подпись для оси х
        plt.ylabel('Температура') #Подпись для оси y
        plt.title('Температурная кривая') #Название
        plt.grid(True)
        plt.show()
```

```python
p1 = Patient(input("Name "), input("Diagnosis "), input("Date of Birth in format YYYYMMDD "))
```
